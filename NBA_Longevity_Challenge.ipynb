##Data Cleaning

```python
traindf.drop_duplicates()
```

```python
X_train = traindf.drop(labels = ["Target"],axis=1)
y_train = traindf["Target"]
X_test = testdf
```

```python
cols = list(traindf.columns)
cols.pop()
```

```python
cols
```

```python
from sklearn import preprocessing
X_train = preprocessing.normalize(X_train)
X_test = preprocessing.normalize(X_test)
```

```python
X_train = pd.DataFrame(X_train, columns = cols)
X_test = pd.DataFrame(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, mean_squared_error
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
```

```python
kfold = StratifiedKFold(n_splits=12)
accuracies = []
```

```python
lmodel = LogisticRegression().fit(X_train,y_train)
knnmodel = KNeighborsClassifier().fit(X_train, y_train)
dctmodel = DecisionTreeClassifier().fit(X_train,y_train)
gnbmodel = GaussianNB().fit(X_train, y_train)
abmodel = AdaBoostClassifier(DecisionTreeClassifier(), learning_rate=0.1).fit(X_train,y_train)
rfmodel = RandomForestClassifier().fit(X_train,y_train)

models = ["Logistic Regression","KNN Classifier","Decision Tree Classifier","GaussianNB Classifier","AdaBoost Classifier","Random Forest Classifier"]
modelscal = [lmodel,knnmodel,dctmodel,gnbmodel, abmodel,rfmodel]
for i in modelscal:
    acc = cross_val_score(i, X_train, y_train, scoring = "accuracy", cv = kfold, n_jobs=4)
    accuracies.append(acc.mean())
    
acc_score = {"Model": models, "Mean Accuracy": accuracies}    
acc_score = pd.DataFrame(acc_score)
acc_score
```

```python
rfmodel= RandomForestClassifier().fit(X_train, y_train)
```

```python
predictions= rfmodel.predict(X_test)
predictions= pd.DataFrame(predictions, columns=['prediction'])
```
